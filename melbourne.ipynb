```python
import pandas as pd
df = pd.read_csv('Melbourne_housing_FULL.csv')
print(df)
```

```python
df.nunique()
#gives unique coloumns
```

```python
df.info()
```

```python
df=df.drop(columns=['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'])
df
```

```python
df=df.dropna()
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,dtype=int)
df
```

```python
#newdf=df.dropna()#delete null value rows
#print(newdf.to_string())
#print(df.to_string())
```

```python
x=df.drop(columns=['Price'])
y=df['Price']
x
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
linear_model.score(x_train,y_train)
```

```python
linear_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
reg2= Lasso(alpha=50,max_iter=100,tol=0.1)
reg2.fit(x_train,y_train)
```

```python
reg2.score(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
reg2= Ridge(alpha=50,max_iter=100,tol=0.1)
reg2.fit(x_train,y_train)
```

```python
reg2.score(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```
